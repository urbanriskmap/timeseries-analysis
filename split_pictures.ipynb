```python
import os
# import chennai_config
# config = chennai_config.config

import jakarta_config
config = jakarta_config.config

from loaders.cognicity_loader import CognicityLoader

loader = CognicityLoader(config)

urls = loader.get_image_urls()

flood = config["flood_pkeys"]
noFlood = config["no_flood_pkeys"]

all_pkeys = config["all_pkeys"]



filelist = os.listdir(loader.image_path)
for filename in filelist:
    print(filename)
    pkey = int(filename.split(".")[0])
    if pkey not in all_pkeys:
        continue
    path = os.path.join(loader.image_path, filename)
    with open(path, "rb") as source:
        if pkey in flood:
            write_path = os.path.join("id_flood", filename)
        else: 
            write_path = os.path.join("id_no_flood", filename)
        with open(write_path, "wb") as dest:
            dest.write(source.read())
```
